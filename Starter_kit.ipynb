## Matplotlib



Matplotlib is a plotting library. In this section we will give a brief introduction to the `matplotlib.pyplot` module, which provides a plotting system similar to that of MATLAB.

By running this special iPython command, we will be displaying plots inline:
`%matplotlib inline`

````
import matplotlib.pyplot as plt
%matplotlib inline
````

### Plotting

````
x = np.linspace(0,10, num = 20)
y_sin = np.sin(x)
y_cos = np.cos(x)

# Plot the points using matplotlib
plt.plot(x, y_sin)
plt.plot(x, y_cos)
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.title('Sine and Cosine')
plt.legend(['Sine', 'Cosine'])
plt.show()
````

#### Subplots

````
# Define our figure. Figsize is given in inches
fig = plt.figure(figsize=(16,16))

# Compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0, 3 * np.pi, 0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)
# Set up a subplot grid that has height 2 and width 1,
# and set the first such subplot as active.
ax = fig.add_subplot(221)
# Make the first plot
ax.plot(x, y_sin)
ax.set_title('Sine')

# Set the second subplot as active, and make the second plot. 
# @param: c = color of the plot 
# @param: linewidth = width of the line
ax = fig.add_subplot(222)
ax.plot(x, y_cos,linewidth=2,c="orange")
ax.set_title('Cosine')


# Third subplot. plot + scatter 
# Marker is the symbol used for scatter plot. Look up here for different types of markers: https://matplotlib.org/3.1.1/api/markers_api.html
ax = fig.add_subplot(223)
ax.plot(x, y_sin/y_cos)
ax.scatter(x, y_sin/y_cos,c="red",marker="s")
ax.set_title('Tangent')

# Co Tanget  
ax = fig.add_subplot(224)
ax.plot(x, y_cos/y_sin)
ax.scatter(x, y_cos/y_sin,c="green",marker="o")
ax.set_title('Cotangent')

# Remove axis
ax.axis("off")

# Show the figure.
plt.show()
````

### Reading image

- Locally the img1.jpg has been downloaded
- Typically in your assignments you will be accessing data from a Google Drive Folder

````
import cv2
%matplotlib inline

# Reading an image in OpenCV
img = cv2.imread('img1.jpg')

# OpenCV reads the image in BGR format. 
# We can change that to RGB using this:
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Printing the shape of the image
print("Rows, Columns, Channels:",img.shape)

# Displaying the image
plt.figure(figsize = [10,10])
plt.imshow(img)
plt.axis('off') # Remove the axes of the plot
plt.show() # To display the image without additional metadata

# Saving an image
cv2.imwrite('saved_image.jpg', img)
````

### Convert Image to Grayscale

Note - You will be implementing these functions in numpy by yourselves later.

````
img_gray = cv2.imread('img1.jpg', 0) # The zero tells OpenCV to read the image as grayscale
plt.figure(figsize = [10,10])
plt.imshow(img_gray,cmap='gray')
plt.axis('off')
plt.show()
````

### Selecting a Region of Interest (Cropping)

````
roi = img[1125:1180, 1280:1500,:]
plt.figure(figsize = [10,10])
plt.imshow(roi)
plt.axis('off')
plt.show()
````

### Resizing an image
You will be asked to implement this from scratch later.

````
resized = cv2.resize(img, (np.int(200*(1920/1280)),200))
plt.axis('off')
plt.imshow(resized)
plt.show()
print("Resized Image size:",resized.shape)
````
